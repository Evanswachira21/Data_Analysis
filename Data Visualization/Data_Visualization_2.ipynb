```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
# Sample Data
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value1': [10, 15, 7, 12, 8],
    'Value2': [5, 10, 3, 8, 4]
})
```

```python
#plt.table?
```

## 1. Simple Text

```python
plt.figure(figsize=(6, 3))
plt.text(0.1, 0.6, '20%', fontsize=100, fontweight='bold', color='#D29576')
plt.text(0.1, 0.4, 'of children had a', fontsize=20, color='#D29576')
plt.text(0.1, 0.3, 'traditional stay-at-home mom', fontsize=20, fontweight='bold', color='#D29576')
plt.text(0.1, 0.2, 'in 2012, compared to 41% in 1970', fontsize=20, color='#D29576')
plt.axis('off')
plt.show()
```

## 2. Scatterplot

```python
# 2. Scatterplot
plt.figure(figsize=(6, 4))
plt.scatter(df['Value1'], df['Value2'])
plt.xlabel("Value1")
plt.ylabel("Value2")
plt.title("Scatterplot")
plt.show()
```

## 3. Table

```python
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
plt.show()
```

```python
# 1. Table with Dark Theme
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center',
                 colColours=['black']*len(df.columns), cellColours=[['grey']*len(df.columns)]*len(df), colWidths=[0.2]*len(df.columns))
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.show()

# 2. Table with Light Grey Theme
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center',
                 colColours=['lightgrey']*len(df.columns), cellColours=[['lightgrey']*len(df.columns)]*len(df), colWidths=[0.2]*len(df.columns))
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.show()

# 3. Table with White Theme
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center',
                 colWidths=[0.2]*len(df.columns))
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.show()
```

## 4. Line Graph

```python
plt.figure(figsize=(6, 4))
plt.plot(df['Category'], df['Value1'], marker='o', label='Value1')
plt.plot(df['Category'], df['Value2'], marker='s', label='Value2')
plt.legend()
plt.title("Line Graph")
plt.show()
```

## 5. Heatmap

```python
data = np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

## 6. Slope Graph

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i in range(len(df)):
    ax.plot([0, 1], [df['Value1'][i], df['Value2'][i]], marker='o')
ax.set_xticks([0, 1])
ax.set_xticklabels(['Value1', 'Value2'])
plt.title("Slope Graph")
plt.show()
```

## 7. Horizontal Bar Chart

```python
plt.barh(df['Category'], df['Value1'])
plt.xlabel("Value")
plt.ylabel("Category")
plt.title("Horizontal Bar Chart")
plt.show()
```

## 8. Vertical Bar Chart

```python
plt.bar(df['Category'], df['Value1'])
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Vertical Bar Chart")
plt.show()
```

## 9. Stacked Vertical Bar

```python
plt.bar(df['Category'], df['Value1'], label='Value1')
plt.bar(df['Category'], df['Value2'], bottom=df['Value1'], label='Value2')
plt.legend()
plt.title("Stacked Vertical Bar Chart")
plt.show()
```

## 10. Stacked Horizontal Bar

```python
plt.barh(df['Category'], df['Value1'], label='Value1')
plt.barh(df['Category'], df['Value2'], left=df['Value1'], label='Value2')
plt.legend()
plt.title("Stacked Horizontal Bar Chart")
plt.show()
```
