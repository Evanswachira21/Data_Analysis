```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample Data
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value1': [10, 15, 7, 12, 8],
    'Value2': [5, 10, 3, 8, 4]
})
```

## 1. Simple Text

```python
plt.figure(figsize=(6, 3))
plt.text(0.1, 0.6, '20%', fontsize=100, fontweight='bold', color='#D29576')
plt.text(0.1, 0.4, 'of children had a', fontsize=20, color='#D29576')
plt.text(0.1, 0.3, 'traditional stay-at-home mom', fontsize=20, fontweight='bold', color='#D29576')
plt.text(0.1, 0.2, 'in 2012, compared to 41% in 1970', fontsize=20, color='#D29576')
plt.axis('off')
plt.show()
```

## 2. Scatterplot

```python
# 2. Scatterplot
plt.figure(figsize=(6, 4))
plt.scatter(df['Value1'], df['Value2'])
plt.xlabel("Value1")
plt.ylabel("Value2")
plt.title("Scatterplot")
plt.show()
```

## 3. Table

```python
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
plt.show()
```

```python
# 1. Table with Dark Theme
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center',
                 colColours=['black']*len(df.columns), cellColours=[['grey']*len(df.columns)]*len(df), colWidths=[0.2]*len(df.columns))
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.show()

# 2. Table with Light Grey Theme
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center',
                 colColours=['lightgrey']*len(df.columns), cellColours=[['lightgrey']*len(df.columns)]*len(df), colWidths=[0.2]*len(df.columns))
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.show()

# 3. Table with White Theme
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center',
                 colWidths=[0.2]*len(df.columns))
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.show()
```

## Q8 

Show how many individual and corporate taxpayers exist in each region using a stacked bar chart.

```python
grouped = df.groupby(by="corporate_taxpayers")['region'].count()

# Convert the grouped result into a DataFrame
df_grouped = pd.DataFrame(grouped).reset_index()

# Set Seaborn style
sns.set_style("darkgrid")

# Create bar plot
ax = sns.barplot(x='corporate_taxpayers', y='region', data=df_grouped, palette="coolwarm")

# Add annotations (show counts on top of bars)
for i, value in enumerate(df_grouped['region']):
    plt.text(i, value + 2, str(value), ha='center', fontsize=12, fontweight='bold')

# Add labels and title
plt.xlabel("Corporate Taxpayers")
plt.ylabel("Number of Regions")
plt.title("Corporate Taxpayer Count by Region")

# Display the plot
plt.show()
```

 Show a bar chart of the top 10 taxpayers based on total tax_paid.

```python


# Grouping and sorting
grouped = df.groupby('taxpayer_id')['tax_paid'].sum()
grouped = grouped.sort_values(ascending=False)

# Select the top 10 taxpayers
top_10_taxpayers = grouped.head(10)

# Reset the index to have 'taxpayer_id' as a column
top_10_taxpayers = top_10_taxpayers.reset_index()

# Merge to get 'taxpayer_name' (assuming 'taxpayer_name' is in df)
top_10_taxpayers = top_10_taxpayers.merge(df[['taxpayer_id', 'taxpayer_name']], on='taxpayer_id', how='left')

# Set Seaborn style
sns.set_style("darkgrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='taxpayer_name', y='tax_paid', data=top_10_taxpayers, palette="viridis")

# Add annotations (show tax_paid value on top of bars)
for i, value in enumerate(top_10_taxpayers['tax_paid']):
    plt.text(i, value + 5000, f'{value:,.0f}', ha='center', fontsize=12, fontweight='bold')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.xlabel("Taxpayer Name")
plt.ylabel("Total Tax Paid")
plt.title("Top 10 Taxpayers by Total Tax Paid")

# Display the plot
plt.show()
```
